```python
import gradio as gr
import os
import shutil
import base64
from datetime import datetime
from io import BytesIO

hasChanged = True

def final_output(prompt):

  global hasChanged

  documents = SimpleDirectoryReader("/content/user_doc").load_data()
  index = VectorStoreIndex.from_documents(documents, service_context=service_context)
  query_engine = index.as_query_engine()
  return query_engine.query(prompt)

def save_pdf(pdf_content, output_folder="user_doc"):
    # Create the directory if it doesn't exist
    os.makedirs(output_folder, exist_ok=True)

    # Generate a filename based on the current datetime
    current_datetime = datetime.now().strftime("%Y%m%d%H%M%S")
    output_filename = os.path.join(output_folder, f"{current_datetime}.pdf")

    # Write the PDF content to the file
    with open(output_filename, "wb") as pdf_file:
        pdf_file.write(pdf_content)

    return output_filename


def my_function(text_input, pdf_file_content=None):
    try:
      if pdf_file_content!="":
          # Generate a PDF file from the extracted text
          pdf_content = base64.b64decode(pdf_file_content)
          return save_pdf(pdf_content)
      if text_input.strip() != "":
          # Use only the text input if no PDF provided
          result = final_output(text_input)
      # Pass the combined text to the query engine
          return result
    except Exception as e:
      return f"Error: {str(e)}"

iface = gr.Interface(
    fn=my_function,
    inputs=[
        "text",
        "text"
    ],
    outputs="text"
)

iface.launch(debug=True)
```
